```python
import networkx as nx
import matplotlib.pyplot as plt

# Example synthetic data
drugs = ['Drug_A', 'Drug_B', 'Drug_C', 'Drug_D', 'Drug_E']
cell_lines = ['CL_1', 'CL_2', 'CL_3', 'CL_4', 'CL_5', 'CL_6']
indications = ['Lung Cancer', 'Breast Cancer', 'Leukemia', 'Melanoma']

# Relationships: Drug tested on Cell line
drug_cell_edges = [
    ('Drug_A', 'CL_1'), ('Drug_A', 'CL_2'),
    ('Drug_B', 'CL_2'), ('Drug_B', 'CL_3'),
    ('Drug_C', 'CL_4'), ('Drug_C', 'CL_5'),
    ('Drug_D', 'CL_1'), ('Drug_D', 'CL_5'),
    ('Drug_E', 'CL_3'), ('Drug_E', 'CL_6')
]

# Relationships: Cell line linked to Indication
cell_indication_edges = [
    ('CL_1', 'Lung Cancer'),
    ('CL_2', 'Lung Cancer'),
    ('CL_3', 'Breast Cancer'),
    ('CL_4', 'Leukemia'),
    ('CL_5', 'Melanoma'),
    ('CL_6', 'Breast Cancer')
]

# Create the tripartite graph
G = nx.Graph()
G.add_nodes_from(drugs, bipartite=0, type='drug')
G.add_nodes_from(cell_lines, bipartite=1, type='cell')
G.add_nodes_from(indications, bipartite=2, type='indication')

# Add edges
G.add_edges_from(drug_cell_edges)
G.add_edges_from(cell_indication_edges)

# Position nodes in three vertical layers
pos = {}
pos.update((node, (0, i)) for i, node in enumerate(drugs))
pos.update((node, (1, i)) for i, node in enumerate(cell_lines))
pos.update((node, (2, i)) for i, node in enumerate(indications))

# Node colors
color_map = []
for node in G.nodes(data=True):
    if node[1]['type'] == 'drug':
        color_map.append('skyblue')
    elif node[1]['type'] == 'cell':
        color_map.append('lightgreen')
    else:
        color_map.append('salmon')

# Draw network
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_color=color_map, node_size=1500, edge_color='gray', font_size=8)
plt.title("Tripartite Network: Drugs – Cell Lines – Indications", fontsize=14)
plt.axis('off')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define three generic node groups
group1 = ['Node_A1', 'Node_A2', 'Node_A3']   # Layer 1
group2 = ['Node_B1', 'Node_B2', 'Node_B3', 'Node_B4']  # Layer 2
group3 = ['Node_C1', 'Node_C2', 'Node_C3']   # Layer 3

# Create edges generically
edges = [
    ('Node_A1', 'Node_B1'), ('Node_A1', 'Node_B2'),
    ('Node_A2', 'Node_B2'), ('Node_A2', 'Node_B3'),
    ('Node_A3', 'Node_B3'), ('Node_A3', 'Node_B4'),
    ('Node_B1', 'Node_C1'), ('Node_B2', 'Node_C1'),
    ('Node_B3', 'Node_C2'), ('Node_B4', 'Node_C3')
]

# Build graph
G = nx.Graph()
G.add_nodes_from(group1, layer=0)
G.add_nodes_from(group2, layer=1)
G.add_nodes_from(group3, layer=2)
G.add_edges_from(edges)

# Position nodes in vertical layers
pos = {}
pos.update((node, (0, i)) for i, node in enumerate(group1))
pos.update((node, (1, i)) for i, node in enumerate(group2))
pos.update((node, (2, i)) for i, node in enumerate(group3))

# Color map by layer
color_map = []
for node in G.nodes(data=True):
    if node[1]['layer'] == 0:
        color_map.append('skyblue')
    elif node[1]['layer'] == 1:
        color_map.append('lightgreen')
    else:
        color_map.append('salmon')

# Draw network
plt.figure(figsize=(8, 5))
nx.draw(G, pos, with_labels=True, node_color=color_map, node_size=1500,
        edge_color='gray', font_size=9)
plt.title("Generic Tripartite Network", fontsize=14)
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Example data -------------------------------------------------------------
labels = ["A", "B", "C", "D"]
pie_data = [
    [35, 25, 25, 15],   # Pie 1
    [30, 20, 35, 15],   # Pie 2
    [20, 30, 30, 20],   # Pie 3
]
# Bar chart will show totals across the three pies for each label
bar_data = np.sum(np.array(pie_data), axis=0)

# --- Figure & subplots --------------------------------------------------------
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
ax1, ax2, ax3, ax4 = axes.flatten()

# --- Styling -----------------------------------------------------------------
plt.rcParams.update({
    "font.size": 11,
    "axes.titleweight": "bold",
})

def pie(ax, sizes, title):
    wedges, texts, autotexts = ax.pie(
        sizes,
        labels=labels,
        autopct="%1.0f%%",
        startangle=90,
        pctdistance=0.8,
        labeldistance=1.05,
        wedgeprops=dict(width=0.45, edgecolor="white")
    )
    # Donut hole
    centre_circle = plt.Circle((0, 0), 0.55, fc="white")
    ax.add_artist(centre_circle)
    ax.set_title(title)
    ax.axis("equal")

pie(ax1, pie_data[0], "Segment Split — Q1")
pie(ax2, pie_data[1], "Segment Split — Q2")
pie(ax3, pie_data[2], "Segment Split — Q3")

# --- Bar chart ----------------------------------------------------------------
x = np.arange(len(labels))
ax4.bar(x, bar_data)
ax4.set_xticks(x, labels)
ax4.set_title("Totals Across Q1–Q3")
ax4.set_ylabel("Total")
ax4.grid(axis="y", linestyle=":", alpha=0.5)

# --- Layout -------------------------------------------------------------------
fig.suptitle("Three Pies + One Bar (Clean Layout)", y=0.98, fontsize=14, fontweight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
